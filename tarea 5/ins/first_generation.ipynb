Notebook para transformar el texto de ultimahoracol a csv con info relevante

```python
import pandas as pd
import seaborn as sns
import unidecode
```

```python
df = pd.read_csv("comorbilidades.csv")
df["comorbilidades"] = df["comorbilidades"].fillna("")
df = df.rename(columns={"comorbilidades": "texto"})
```

```python
df.head()
```

```python
similaridades = {
    "hta":["hipertension", "hipertensión", "hipertenso", "hta"],
    "obesidad": ["obesidad", "obeso"],
    "fumador": ["fumador", "exfumador", "tabaquismo", "extabaquismo"],
    "hipoxia":["hipoxia"],
    "cardiovascular": ["cardiovascular", "cardiopatia", "cardiaca", "circulatorias", "trombosis", "anticuag", "coronaria"],
    "diabetes": ["diabetes"],
    "enfermedad respiratoria": ["asma", "epoc", "neumonia", "oxigeno"],
    "enfermedad renal": ["renal"],
    "hipotiroidismo": ["hipotiroidismo"],
    "cancer": ["cancer", "masa", "leucemia", "tumor"],
    "alcoholismo": ["alcoholismo", "alcohol"]
    
}
def estandarizar(s):
    s = s.lower()
    s = unidecode.unidecode(s)
    s = s.replace(",","")
    s = s.replace(".","")
    s = s.replace(":","")
    s = s.replace("comorbilidades","")
    return s

def generar_lista(s):
    lista = ""
    for sim in similaridades:
        for opt in similaridades[sim]:
            if opt in s:
                lista += f"{sim},"
                break
    if lista != "":
        lista = lista[:-1]
    return lista
```

```python
df["texto_comorb"] = df["texto"].apply(estandarizar)
df["comorbilidades"] = df["texto_comorb"].apply(generar_lista)
df.to_csv("comorbilidades2.csv", sep=";", index=False, columns=["fecha", "genero", "edad", "ciudad", "texto_comorb", "comorbilidades"])
```
