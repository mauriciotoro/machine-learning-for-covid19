```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.style.use('https://raw.githubusercontent.com/davidbetancur8/palettes/master/eafit.mplstyle')
enmax_palette = ["#000066", "#00A9E0", "#FF8F1C", "#C4D600", "#00B0B9"]

sns.set_palette(palette=enmax_palette)
```

```python
df_confirmed = pd.read_csv("Confirmed.csv")
```

```python
df_confirmed.head()
```

```python
df_col = df_confirmed[df_confirmed["Country"] == "Colombia"]
df_col["conf_ewma"] = df_col["Confirmed"].ewm(com=0.5).mean()
```

```python
df_col.head()
```

```python
df_col.plot(x="Date", y=["Confirmed", "conf_ewma"], figsize=(10,5), linewidth=5)
plt.xticks(rotation=90)
plt.savefig(f'emwa.png',bbox_inches='tight', transparent="True", pad_inches = 0)

plt.show()
```
