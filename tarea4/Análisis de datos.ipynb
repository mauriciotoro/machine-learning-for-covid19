```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
def limpieza(dataFrame):
    tablaLimpia = pd.read_csv(dataFrame, header=0)
    tablaLimpia = tablaLimpia.drop(['Unnamed: 33', 'Unnamed: 34', 'Unnamed: 35','Unnamed: 36','Unnamed: 37','Unnamed: 38','Unnamed: 39'], axis=1)
    tablaLimpia = tablaLimpia.drop(['Unnamed: 40', 'Unnamed: 41', 'Unnamed: 42','Unnamed: 43','Unnamed: 44'], axis=1)
    resultado = recorrerColumnasTabla(tablaLimpia)
    return resultado

def recorrerColumnasTabla(tablaLimpia):
    tipos = tablaLimpia.dtypes
    for i in range(len(tablaLimpia.columns)):
        nombreColumna = tablaLimpia.columns[i]
        nombreNuevoColumna = nombreColumna.upper().strip().replace(" ", "")
        tablaLimpia = tablaLimpia.rename(columns={nombreColumna: nombreNuevoColumna})  # opcional
        if tipos[i] == 'O' and nombreNuevoColumna!='AGE':
            arregloPalabras = separarLetras(pd.unique(tablaLimpia[nombreNuevoColumna]))
            tablaLimpia[nombreNuevoColumna] = quitarVariables(arregloPalabras,tablaLimpia,nombreNuevoColumna,pd.unique(tablaLimpia[nombreNuevoColumna]))
    return tablaLimpia

def separarLetras(columna):
    arregloLetras= []
    letra = ""
    for i in range(len(columna)):
        numeroenColumna = str(columna[i])
        if not(numeroenColumna=="nan"):
            numeroenColumna = numeroenColumna.upper().strip()
            if not(numeroenColumna in arregloLetras):
                arregloLetras.append(numeroenColumna)
    return arregloLetras

def quitarVariables(arregloNuevo,tablaLimpia,columna,arregloViejo):
    tablaSolucion = tablaLimpia[columna]
    for i in range(len(arregloNuevo)):
        for j in range(len(arregloViejo)):
            if( str(arregloViejo[j]).upper().strip() == str(arregloNuevo[i])):
                nuevaColumna = tablaSolucion.replace(str(arregloViejo[j]), str(arregloNuevo[i]))
                tablaSolucion = nuevaColumna 
    return tablaSolucion
```

```python
tablaFinal = limpieza('COVID19_open_line_list.csv')
tablaFinal = tablaFinal.drop([5873],axis=0)
tablaFinal = tablaFinal[0:13173]
```
